```python
import pandas as pd
from string import punctuation
import re

def remove_normal_characters(text):
    text.encode('ascii', 'ignore').decode("utf-8")
    return re.sub(r'[^{}]+'.format(punctuation),'',text)

df = pd.read_csv('Structured_DataFrame_Main_Categories.csv', index_col=0)
df['Item Description'] = df['Item Description'].apply(lambda d: remove_normal_characters(str(d)))

df
```
