# 23. Mapping the new Dataset
We want to have a mapped dataset of the new data to the Interpol topics list.

```python
import pandas as pd
import pickle

filenameCategories = "../darkweb/data/pickle/categorieen.pkl"

labels = pickle.load(open(filenameCategories, 'rb'))

d = { 'Label': labels }
df = pd.DataFrame(d)
for i, row in df.iterrows():
    if row.Label == 'Weapons':
        df.at[i, "Label"] = "Weapons/abcd"

labels
```

```python
df2 = pd.read_csv("../Data/interpoltopics.csv").dropna()
df2
```

```python
def edit(category):
    print(category)
    for i, row in df2.iterrows():
        if category in row["Agora"]:
            return row["Interpol"]
```

```python
df["Label"] = df["Label"].apply(lambda d: edit(d))
```
