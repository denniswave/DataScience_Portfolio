# 2. Bag of Words
After I was done with the preprocessing, I looked into a method for vectorization. Other group members started with Word2Vec and TFIDF. I used the kaggle dataset and preprocessed with the previously written preprocessing script, imported from a seperate file.

```python
import pandas as pd
from preprocessing import PreProcessor

pp = PreProcessor()

df = pd.read_csv('darkweb/data/agora.csv')

descriptions = df[' Item'] + " " + df[' Item Description']
ppDescriptions = descriptions[:1000].apply(lambda d: pp.preprocess(str(d)))
ppDescriptions
```

## Vectorizing the descriptions
After the descriptions are preprocessed, they can be transformed into vectors. I used sklearn for this and transformed the first 1000 descriptions into vectors. Bag of Words creates a vector with a length equal to the sum of unique words in all descriptions combined. In our dataset, there are over 109.000 descriptions and the unique word count will be quite high. Therefore, when I tried to create a Bag of Words with all descriptions, I got a memory error. Because of this, I only took the first 1000 descriptions, preprocessed those and used those to create a Bag of Words. 

You can see how a Bag of Words is succesfully created. For every appearance of a word in a description, the count for that word is reflected in the Bag of Words. For example: you can see that the word 'account' appears twice in the third description. 

Out of the 1000 descriptions (rows), come 1963 columns: unique words. Because this approach can not be used with to many words, we decided probably not to use this in our further work on the project. 

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()

bag_of_words = count_vectorizer.fit_transform(ppDescriptions)

feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```
